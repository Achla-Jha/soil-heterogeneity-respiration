# Code for exponential relationship of oxygen with the aggregate size; a and b are the empirical parameters

```python
def exponential(x, a, b):
    return a*np.exp(b*x)
```

# Code for power law relationship of MB and C with the aggregate size; a and b are the empirical parameters

```python
def power_law(x, a, b):
    return a*np.power(x, b)
```

# Conceptual plot for depicting the relationship of MB, C and Oxygen with aggregate size distribution at a particular soil moisture content

```python
fig = plt.figure(figsize=(6.5, 5.5))
ax1 = fig.add_subplot(111) 
fig.canvas.draw()
x = np.linspace(0,10,100)
x1 = np.linspace(0,3.9,100)
x6 = np.linspace(3.9,9.2,100)
x2 = np.linspace(0,10,100)
x3 = np.linspace(0,10,100)
y2 = stats.gamma.pdf(x,a=4.5, scale=1) # a is alpha and scale is the beta
y3 = exponential(x1, 0.0012, 1.10)
y6 = exponential(x6, 0.040, 0.20)
y4 = power_law(x2, 0.1, 0.35)
y5 = power_law(x2, 0.12, 0.28)
ax1.plot(x, y2)
ax1.plot(x1, y3,color='green')
ax1.plot(x6, y6,color='green')
ax1.plot(x2, y4,linestyle="--",color='red')
ax1.plot(x3, y5,linestyle='-.',color='purple')
ax1.text(0.7,0.154,'[MB]')
ax1.text(0.75,0.08,'[C]')
ax1.text(1,0.02,'[O$_{2}$]')
ax1.text(3,0.09,'D$_{w}$')
ax1.text(8,0.18,'D$_{o}$')
ax1.spines['top'].set_visible(False)
ax1.spines['right'].set_visible(False)
ax1.fill_between(x, y2, 0, where = (x > 0) & (x <= 4), color='blue', alpha=.1)
labels = [item.get_text() for item in ax2.get_xticklabels()]
labels[1] = 0
labels[2] = ''
labels[3] = '$R_{a}^{w}$'
labels[4] = ''
labels[5] = ''
labels[6] = '$R_{a}$'
plt.arrow(4.0, 0.220, -1.1, 0, head_width=0.03, width = 0.017, head_length=0.6, facecolor='lightsteelblue')
plt.text(0.18, 0.250, 'Aggregates surrounded', color='red', fontsize = 12)
plt.text(0.2, 0.238, 'by water-filled pores', color='red', fontsize = 12)

ax1.set_xticklabels(labels)
ax1.set_xlabel(r'Aggregate size')
ax1.yaxis.set_tick_params(labelleft=False)
ax1.set_yticks([])
plt.show()
```
